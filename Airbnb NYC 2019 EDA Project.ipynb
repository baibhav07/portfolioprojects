```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading the dataset into the notebook
df=pd.read_csv('Airbnb NYC 2019.csv')
```

```python
# checking the head of the dataset
df.head()
```

```python
df.info()
```

```python
# checking the total number of null values
df.isnull().sum()
```

```python
df.describe()
```

```python
df['last_review']
   
```

```python
# replacing reviews per month column with mean 
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].mean())
```

```python
df.isnull().sum()
```

```python
# dropping columns that are not useful
df.drop('last_review',axis=1,inplace=True)
```

```python
df.info()
```

```python
# dropping null values using indexes

name_null=df[df['name'].isnull()].index
```

```python
df.drop(name_null,axis=0,inplace=True)
```

```python
host_name_null= df[df['host_name'].isnull()].index
```

```python
df.drop(host_name_null,axis=0,inplace=True)
```

```python
Q1 = 69
Q3 = 175
IQR = Q3 - Q1
```

```python
IQR 
```

```python
upper_limit = Q3+1.5*IQR
lower_limit = Q1-1.5*IQR
```

```python
# outlier detection using IQR

len(df[df['price']>=upper_limit])
```

```python
# percentage of values in outliers

len(df[df['price']>=upper_limit])*100/len(df)
```

```python
upper_limit
```

```python
outlier_index=df[df['price']>=upper_limit].index
```

```python
df.drop(outlier_index,axis=0,inplace=True)
```

```python
# most expensive listings

df.groupby(['name','neighbourhood','neighbourhood_group','room_type'])[['price']].mean().sort_values(by='price',ascending=False).head()
```

Manhattan also is the most expensive borough given that its the most popular among people and has the maximum no of listings.

A large share of the high priced listings are situated in Manhattan, followed by Brooklyn.
